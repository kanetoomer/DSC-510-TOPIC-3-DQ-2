```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# My parameters
true_p = 0.5 # I want it to be fair
n_trials = 1000  # My fingers hurt after all of these coin flips lol
confidence_level = 0.95
z = norm.ppf((1 + confidence_level) / 2)

# Simulating the coin flips
flips = np.random.binomial(1, true_p, n_trials)

# Estimating p as a function of the number of flips
estimates = np.cumsum(flips) / np.arange(1, n_trials + 1)

# Calculating the margin of error
margins_of_error = z * np.sqrt(estimates * (1 - estimates) / np.arange(1, n_trials + 1))

# Plotting the estimates and confidence intervals
plt.figure(figsize=(10, 6))
plt.plot(estimates, label='Estimated p')
plt.fill_between(range(n_trials), estimates - margins_of_error, estimates + margins_of_error, color='b', alpha=0.2, label='Confidence Interval')
plt.axhline(true_p, color='r', linestyle='--', label='True p (hypothetical)')
plt.xlabel('Number of Flips')
plt.ylabel('Estimated Probability')
plt.title('Estimation of Probability of Heads (Fair Coin)')
plt.legend()
plt.show()
```
